```python
from pprint import pprint

import pickle
import json

import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
with open('saved-data/python/car_list.pickle', 'r') as f:
    car_list = pickle.load(f)
```

```python
def getReview(car_info):
    all_review = []
    
    
    base_url = 'http://www.edmunds.com/'
    base_url = base_url + \
          '/'.join([car_info['make'], car_info['model'], car_info['year'], car_info['type']])
            
    count = 0
    while True:
        count += 1
        print 'count: ', count
        
        url = base_url + '/consumer-reviews/pg-' + str(count) + '/'
        
        print 'url: ', url
        
        response = requests.get(url)
        review_soup = BeautifulSoup(response.content, 'lxml')
        if response.ok == True:
            try:
                widget = review_soup.find('div', id='body-reviews-widget')
                container = widget.findAll('div', class_='individual-review-container')

                print 'container length: ', len(container)

                if len(container) == 0:
                    break
                else:
                    for c in container:
                        header = c.findAll('h3')
                        text = c.findAll('p', class_='review-text')
                        single_review = {}
                        for i in range(len(header)):
                            single_review[header[i].text] = text[i].text

                        all_review.append(single_review)
            except:
                print 'exception'
                break
        else:
            print 'page unvailable'
            break
    
    if all_review:
        return all_review
```

```python
result = []
for i in range(700, len(car_list)):
    car = car_list[i]
    if car['year'] != '2017':
        val = {}
        val['car descriptiion'] = {}
        val['car descriptiion']['make'] = car['make']
        val['car descriptiion']['model'] = car['model']
        val['car descriptiion']['year'] = car['year']
        val['car descriptiion']['type'] = car['type']

        val['car review'] = getReview(car)

        result.append(val)
    
```

```python
count = 0
for review in result:
    if not review['car review']:
        count += 1
print count
```
